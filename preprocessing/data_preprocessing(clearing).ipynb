```python
import pandas as pd
import os

# 파일 경로 설정 (필요시 수정)
folder_path = './'  # 파일들이 저장된 디렉터리

# 모든 파일명 생성
file_names = [f'C:/Users/user/Desktop/gg/output_data/식품안전_{i}_{i * 50000}.csv' for i in range(1, 20)]

# 데이터프레임들을 저장할 리스트
dataframes = []

# 파일을 순차적으로 읽어 리스트에 추가
for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    try:
        df = pd.read_csv(file_path)
        dataframes.append(df)
        print(f'Loaded {file_name}, shape: {df.shape}')
    except Exception as e:
        print(f'Error reading {file_name}: {e}')

# 데이터프레임들을 하나로 병합
merged_df = pd.concat(dataframes, ignore_index=True)

# 결과 저장
merged_df.to_csv('식품안전_통합.csv', index=False)
print(f'병합 완료. 총 {merged_df.shape[0]}행 저장됨.')
```

# 데이터 제거(제공하지 않아도 되는 제품들 Filtering)
 * 해당 키워드 포함되지 않은 제품들만 남도록 필터링

```python
df[~df["제품명"].str.contains("맥주효모|비오틴", case=False, na=False)]
```

```python
# 제거할 키워드 목록
keywords = ['막걸리향', '새로단', '리포좀', '글루타치온', '환', '병품', '히알루론산', '멜라토닌', '밀크씨슬', '향료', '덴탈', '난각막', '구강',
            '락토페린', '저분자', '콘드로이친',  '글리칸', '이노시톨', '연골', '수출용', '모닝파파', '레티놀', '콜라겐', '아연', '여성유산균', '두피', 
            '엔자임', '루테인', '오메가']

# 정규표현식 패턴 만들기 (예: '효소|베이스|향료')
pattern = '|'.join(keywords)

# 제품명에 해당 키워드가 포함되지 않은 행만 남기기
df = df[~df['제품명'].str.contains(pattern, na=False)].reset_index(drop=True)
df.info()
```

```python
# 제품명에 해당 키워드가 포함되지 않은 행만 남기기
df = df[~df['식품대분류명'].str.contains('동물성가공식품류', na=False)].reset_index(drop=True)
df.info()
```

```python
# 파일로 저장
df.to_csv('C:/Users/user/Desktop/zerosweet/output_data/preprocessing_data/Ver.05/refined_with_gi_non_duplicate(big_class).csv', encoding='utf-8-sig', index=False)
```

# 상위 업체 데이터 파일 생성

```python
# 업체명 빈도수 출력
df = pd.read_csv('refined_with_gi_non_duplicate(big_class).csv', encoding = 'utf-8-sig')

df
```

```python
df_office = df['업체명'].value_counts()
df_office
```

```python
df_office = df['업체명'].value_counts().reset_index()

# 컬럼명 변경
df_office.columns = ['업체명', '제품수']

# 순위 추가
df_office['순위'] = df_office['제품수'].rank(method='dense', ascending=False).astype(int)

# 보기 좋게 정렬
df_office = df_office.sort_values('순위')

df_office
```

```python
df_office.to_csv('office_value_count.csv', encoding='utf-8-sig', index=False)
```
